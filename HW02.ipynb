# Selection of Data and Reshaping the Data

Remove all information that we do not need (keep only information related to Binge Drinking and Poverty)

```python
BingeDrinking_Poverty = CDI[(CDI['Question'] == 'Binge drinking prevalence among adults aged >= 18 years') | (CDI['Question'] == 'Poverty')]
```

```python
BingeDrinking_Poverty.head()
```

```python
GenderStratification = BingeDrinking_Poverty[(BingeDrinking_Poverty['Stratification1'] == 'Female') | (BingeDrinking_Poverty['Stratification1'] == 'Male')]
```

```python
GenderStratification.head()
```

```python
GenderStratification.shape
```

```python
GenderStratification_long = GenderStratification[['LocationAbbr', 'LocationDesc','YearStart', 'Question', 'Stratification1', 'DataValue']]
GenderStratification_long
```

```python
GenderStratification_long['binge_all'] = np.where(GenderStratification_long['Question']!= 'Poverty', True, False)
GenderStratification_long.head()
```

```python
GenderStratification_long['poverty']=np.where(GenderStratification_long['Question'] == 'Poverty', True, False)
GenderStratification_long
```

```python
GenderStratification_long['binge_male_info']=GenderStratification_long['Question']+GenderStratification_long['Stratification1']
GenderStratification_long
GenderStratification_long['binge_male'] = np.where(GenderStratification_long['binge_male_info']=='Binge drinking prevalence among adults aged >= 18 yearsMale', True, False)
GenderStratification_long
```

```python
GenderStratification_long['binge_female_info']=GenderStratification_long['Question']+GenderStratification_long['Stratification1']
GenderStratification_long
GenderStratification_long['binge_female'] = np.where(GenderStratification_long['binge_female_info']=='Binge drinking prevalence among adults aged >= 18 yearsFemale', True, False)
GenderStratification_long
```

```python
binge_clean1=GenderStratification_long.rename(columns = {'LocationAbbr' : 'stateabb', 'LocationDesc' : 'state', 'YearStart':'year'})
binge_clean1.head()
```

```python
binge_clean1.shape
```

```python
binge_clean=binge_clean1[['state','stateabb','year','binge_all','binge_male','binge_female','poverty']]
binge_clean
```

```python
binge_clean.to_csv("binge_clean.csv")
```
